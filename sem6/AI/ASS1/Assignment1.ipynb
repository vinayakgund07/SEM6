```python
#Q1:bfs
from collections import deque

graph = {
        'A':['B','C'],
        'B':['D'],
        'C':['F'],
        'D':['G'],
        'F':[],
        'G':[]
}

def bfs(start):
    visited=[]
    queue = deque([start])
    while queue:
        node = queue.popleft()
        if node not in visited:
            visited.append(node)
            queue.extend(graph[node])
    return visited

result=bfs('D')
print("BFS Traversal starting from node: A")
print(result)
```

```python
#Q2:dfs

graph={
    'root':['Folder1','Folder2'],
    'Folder1':[],
    'Folder2':[]
}

visited=[]

def dfs(node):
    visited.append(node)
    for child in graph[node]:
        dfs(child)
    return visited

result = dfs('root')

print("Dfs traversal: ")
print(result)
```

```python
#B1:A* algorithm

from heapq import heappush,heappop

graph = {
    'S':{'A':1,'B':4},
    'A':{'D':5},
    'B':{'G':1},
    'D':{'G':3},
    'G':{}
}

h={
    'S':7,
    'A':6,
    'B':5,
    'D':2,
    'G':0
}

def astar(start,goal):
    pq=[]
    heappush(pq,(h[start],0,start,[start]))
    
    while pq:
        f,g,node,path = heappop(pq)
        
        if node == goal:
            return path,g
        
        for neighbor,cost in graph[node].items():
            new_g = g + cost
            new_f = new_g + h[neighbor]
            heappush(pq,(new_f,new_g,neighbor,path+[neighbor]))
            
path,cost = astar('S','G')
print("Opitimal Path: ",path)
print("Total Cost: ",cost)
```

```python
#B2:best first search                                                                                                                                                  

from heapq import heappush,heappop

graph = {
    'Start':['X','Y'],                                                                                                                                                                                                                                                  
    'x':['X','Y'],
    'Y':['Z'],
    'Z':['T'],
    'T':[],
}

h={
    'Start':7,
    'X':6,
    'Y':5,
    'Z':2,
    'T':0
}

def best_first_search(start,goal):
    pq=[]
    heappush(pq,(h[start],start))
    visited = []
    while pq:
        cost,node = heappop(pq)
        visited.append(node)
        
        if node == goal:
            return visited 
        
        for neighbour in graph[node]:
            heappush(pq,(h[neighbour],neighbour))
            
result = best_first_search('Start','T')

print("Best First Search Traversal: ")
print(result)
```

```python
#B3:water jug
def water_jug():
    queue=[(0,0,[])]
    visited=set()
    while queue:
        j1,j2,path=queue.pop(0)
        if(j1,j2) in visited:
            continue
        visited.add((j1,j2))
        new_path=path+[(j1,j2)]
        if j1==2 and j2==0:
            matrix=[list(x) for x in new_path]
            return matrix
        moves=[(4,j2),(j1,3),(0,j2),(j1,0),(max(0,j1-(3-j2)),min(3,j1+j2)),(min(4,j1+j2),max(0,j2-(4-j1)))]
        for m in moves:
            queue.append((m[0],m[1],new_path))
print("Water jug steps:",water_jug())
```
