```python
def chatbot():
    print("STUDENT HELP BOT (type byee to exit)")
    while True:
        user=input("You").lower()
        if user=="bye":
            print("bot:Goodbye!")
            break
        elif "fees" in user:
             print("bot:Fees details are on college portal")
        elif "timetable" in user:
              print("bot: Exam start from next month,check acadamic calender.")   
        elif "contact" in user:
              print("bot: You can contact admin at admin@college.edu")
    else:
              print("bot:Sorry, i did'nt understand ..please ask something else!!")
chatbot()
               
```

```python
def self_driving_system(light,obstacle):
    if light=="red":
        return "stop"
    elif light=="yellow":
         return "slow down"
    elif light=="green":
         if obstacle:
             return "Avoid obstacles"
         else :
              return "move"
light=input("Traffic light(red/yellow/green:)")
obstacle=input("obstacle ahed(yes/no)")
result=self_driving_system(light,obstacle=="yes")
print("action",result)
```

```python
def diagnosis():
    fever=input("fever?(yes/no):").lower()
    cough=input("cough?(yes/no):").lower()
    pain=input("body pain?(yes/no):").lower()
    if fever=="yes"and cough=="yes":
        print("diagnosis:common cold")
    elif fever=="yes"and pain=="yes":
        print("diagnosis:flu")
    elif pain=="yes":
        print("diagnosis:muscle strain")
    else:
        print("diagnosis:healthy")
diagnosis()
```
