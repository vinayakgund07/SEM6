```python
# Smart Home System using user input

time = input("Is it Evening (Yes/No):")
presence = input("Is someone at home (Yes/No):")
temperature = int(input("Enter the temperature in degree :"))
clock = input("Is it after 9pm (Yes/No):")
print("\nActions")

if time == "Yes" and presence == "Yes":
    print("Action : Turn on Lights")
if temperature < 18:
    print("Action : Turn on Heating")
if clock == "Yes":
    print("Action : Turn on Thermostat")
    
```

```python
fever = input("Do you have fever (Yes/No):")
cough = input("Do you have cough (Yes/No):")
headache = input("Do you have headache (Yes/No):")
chest_pain = input("Do you have chest_pain(Yes/No):")
print("\nActions")

if fever == "Yes" and headache == "Yes":
    print("Diagnosis :Flu")
if fever == "Yes" and headache == "Yes" and chest_pain == "Yes":
    print("Diagnosis :Pneumonia")
if cough == "Yes" and headache == "Yes":
    print("Diagnosis :Common Cold")
```
